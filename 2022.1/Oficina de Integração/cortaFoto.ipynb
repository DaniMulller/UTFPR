```python
import matplotlib.pyplot as plt
import matplotlib.image as img
from skimage import io
from skimage.util import crop
from PIL import Image

foto = io.imread('https://i.ibb.co/87WcDrZ/Whats-App-Image-2022-06-30-at-11-50-28.jpg')

#cortar 80 pixels em y da foto original
cropped = crop(foto, ((0, 80), (0, 0), (0,0)), copy=False)

print(foto.shape, cropped.shape)

plt.figure(figsize=(20,10))
plt.subplot(121), plt.imshow(foto), plt.axis('off') 
plt.subplot(122), plt.imshow(cropped), plt.axis('off') 
plt.show()
```
